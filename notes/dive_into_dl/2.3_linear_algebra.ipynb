When traveling between two points in downtown manhattan, the distance you need to cover in terms of coordinates is streets horizontal and streets vertical. You cannot travel diagonally.

```python
X = torch.rand(2,3,4)

print(X.sum(0).shape) # (3,4)

print(X.sum(1).shape) # (2, 4)

print(X.sum(2).shape) # (2, 3)
```

```python
# Computes the scalar magnitude.
torch.linalg.norm(X)
```

If you compute (AB)C, you end up with an additional (10, 5) matrix, then an additional (10, 14) matrix.

If you compute A(BC), you end up with an additional (16, 14) matrix, then an additional (10, 14) matrix.

So it's more efficient to compute (AB)C.

```python
A = torch.rand(100, 200)
B = torch.rand(100, 200)
C = torch.rand(100, 200)

A_B_C = torch.stack([A, B, C])

B_slice = A_B_C[1, :, :]

print(B == B_slice)
```
