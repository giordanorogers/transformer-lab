To know which functions and classes can be called in a module, we invoke the `dir` function.

```python
import torch

dir(torch.random)
```

Generally, we can ignore functions that start and end with __ (special objects in Python) or functions that start with a single _ (usually internal functions).

For specific instructions on how to use a given function or class, we can invoke the help function.

```python
help(torch.ones)
```
